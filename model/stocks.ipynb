```python
import yfinance as yf
import pandas as pd

# Define the stock tickers based on the table provided
tickers = {
    "Apple": "AAPL",
    "HSBC": "HSBC",
    "Pepsi": "PEP",
    "Tencent": "TCEHY",
    "Toyota": "TM"
}
start_date = "2015-01-01"
end_date = "2025-11-09" # Using today's date as the end date

# Download historical closing prices for all tickers at once
print(f"Downloading stock data for: {list(tickers.keys())}...")
try:
    stock_data = yf.download(list(tickers.values()), start=start_date, end=end_date, progress=False)
    
    # We are interested only in the 'Close' prices
    close_prices = stock_data['Close']
    
    # Rename the columns to match the names in your table
    close_prices.columns = list(tickers.keys())
    
    # Calculate the summary statistics using the .describe() method
    summary_stats = close_prices.describe()
    
    # Rename the index to match your table's format for clarity
    summary_stats.rename(index={
        'std': 'Std Dev',
        '25%': '25th Percentile',
        '50%': 'Median',
        '75%': '75th Percentile'
    }, inplace=True)
    
    # Display the resulting table
    print("\nSummary Statistics of Stock Prices:")
    display(summary_stats)

except Exception as e:
    print(f"\nAn error occurred. This might be due to a network issue or an invalid ticker.")
    print(f"Error details: {e}")
```
